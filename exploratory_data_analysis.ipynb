The data-set aims to answer the following key questions:

3. How does Infant and Adult mortality rates affect life expectancy?

5. What is the impact of schooling on the lifespan of humans?

```python
import pandas as pd
import numpy as np 
import seaborn as sns 
import matplotlib.pyplot as plt
```

## Read the Dataset 

```python
dataset = pd.read_csv("Life Expectancy Data.csv")
```

```python
#visualizing our target which is the life expentancy 
target = 'Life expectancy '
fig, axes= plt.subplots(1,2, figsize = (12,8))
sns.histplot(x = target, data = dataset, alpha = 0.3, ax = axes[0])
axes[0].set_title('histogram of life_expentancy',fontsize = 14)
sns.kdeplot(x=target,data = dataset ,alpha = 0.3,ax =axes[1],fill=True)
axes[1].set_title("density plot for life expentancy",fontsize=14)
plt.show()
fig.savefig('dist_life_expectancy.jpeg',dpi=82,transparent = False)
```
